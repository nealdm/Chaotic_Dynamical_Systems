```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
D = lambda x: 2*x % 1     # D is the doubling function

def compose(f,n,x):
    '''
    Takes in a function (fun), a desired number of iterations (n), and 
    an initial value (x), and returns the nth composition of the function
    applied to x.
    
    Attributes:
    f (function)
    n (int)
    x (float or whatever value type f can take in)
    
    Returns:
    f^n(x) (float or whatever value type f returns)     
    '''
    x0 = x # make a copy of x just to be safe (shouldn't be a problem, but just in case)
    for i in range(n):
        x0 = f(x0)
    return x0    
```

```python
n = 8

x_space = np.linspace(0,1,1000000) # create a linspace for x to visualize
                                  # the doubling function
y_space = compose(D,n,x_space)    # create a corresponding y linspace with th
                                  # the desired number of iterations
plt.plot(x_space,y_space,'b,')     # plot the graphic with ',' or '.' (pixels or points)
                                  # so there are no vertical line artifacts from
                                  # the plotter trying to make the whole thing connected
plt.plot(x_space,x_space,'k')
plt.rcParams["figure.figsize"] = [30,30]
plt.rcParams.update({'font.size': 40})
plt.title(f"If this were plotted correctly you should see  {2**n} lines")
plt.show()
```
